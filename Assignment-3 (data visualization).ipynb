# Data Visualization on titanic data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

# Line Plot

```python
plt.plot(df.age,df.sibsp)
```

```python
plt.plot(df.pclass ,df.age)
```

```python
from numpy import sin
from numpy import cos
x = [x*0.1 for x in range(100)]
y = sin(x)
z = cos(x)
plt.plot(x,y,'r^',linewidth = 4)
plt.plot(x,z,'go',linewidth = 4)
plt.show()
```

```python
x = df.survived
y = df.pclass
plt.plot(x,y,'go',linewidth = 4)
plt.xlabel('suvived')
plt.ylabel('pclass')
plt.title('Simple Line plot',color = 'darkblue',fontsize = 40)
```
